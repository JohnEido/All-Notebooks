```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Gathering Data using pandas
dataset = pd.read_csv('Q1_data.csv')
```

```python
#Checking the contents of the Data
dataset.head()
```

```python
#Preparing Data
#Removing uneeded and null data
dataset = dataset.drop(columns='id', axis=1)
dataset = dataset.drop(columns='Unnamed: 32', axis=1)
```

```python
dataset.head()
```

```python
#Reassigning Diagnonsis values from string to integer
dataset['diagnosis'] = dataset['diagnosis'].map({'M':1,'B':0})
```

```python
dataset.head()
```

```python
#Examining all feature labes
dataset.columns
```

```python
#Assigning features to their distinct group
features_mean = list(dataset.columns[1:11])
features_se = list(dataset.columns[11:21])
features_worst = list(dataset.columns[1:11])
```

```python
#Removing diagnosis from features
features_mean.append('diagnosis')
features_se.append('diagnosis')
features_worst.append('diagnosis')
```

```python
#Correlating features
correlate = dataset[features_mean].corr()
correlate
```

```python
correlate = dataset[features_se].corr()
correlate
```

```python
correlate = dataset[features_worst].corr()
correlate
```

```python
pred_vars = ['radius_mean','perimeter_mean','area_mean','compactness_mean','concavity_mean','concave points_mean','radius_se','area_se','radius_worst','perimeter_worst','compactness_worst']
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(dataset, test_size = 0.15, random_state=1)
```

```python
train_x = train[pred_vars]
train_y = train['diagnosis']
test_x= test[pred_vars]
test_y = test['diagnosis']
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(train_x,train_y)
```

```python
predictions = model.predict(test_x)
test_y
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, predictions)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
```

```python
precision = precision_score(test_y, predictions)
print("Precision is %.2f" % precision)
recall = recall_score(test_y, predictions)
print("Recall is %.2f" % recall)
accuracy = accuracy_score(test_y, predictions)
print("Accuracy is %.2f" % accuracy)
```
